```python
import os

os.makedirs("/content/raw", exist_ok=True)
os.makedirs("/content/processed", exist_ok=True)
os.makedirs("/content/output", exist_ok=True)

import shutil
shutil.move("customer_churn_dataset-testing-master.csv", "/content/raw/customer_churn.csv")
```

```python
import pandas as pd

df = pd.read_csv("/content/raw/customer_churn.csv")

print(df.shape)
df.head()
```

```python
df = df.drop_duplicates()

df = df.fillna(0)

df.columns = df.columns.str.lower().str.replace(" ", "_")

if 'tenure' in df.columns:
    df['tenure'] = df['tenure'].astype(int)

if 'monthlycharges' in df.columns:
    df['monthlycharges'] = df['monthlycharges'].astype(float)

if 'monthlycharges' in df.columns and 'tenure' in df.columns:
    df['total_charges_calc'] = df['monthlycharges'] * df['tenure']

if 'churn' in df.columns:
    df['churn_flag'] = df['churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
print(df.columns.tolist())
```

```python
customers = df[['customerid','age','gender','churn_flag']]

services = df[['customerid','subscription_type','contract_length',
               'usage_frequency','support_calls','payment_delay']]

billing = df[['customerid','tenure','total_spend','last_interaction']]
```

```python
customers.to_csv("/content/processed/customers.csv", index=False)
services.to_csv("/content/processed/services.csv", index=False)
billing.to_csv("/content/processed/billing.csv", index=False)
```

```python
import sqlite3

conn = sqlite3.connect("/content/output/database.sqlite")

customers.to_sql("customers", conn, if_exists="replace", index=False)
services.to_sql("services", conn, if_exists="replace", index=False)
billing.to_sql("billing", conn, if_exists="replace", index=False)

conn.close()
```
